```python
import pandas as pd
bases = pd.read_csv("../../data/bases.csv")
ranges = pd.read_csv("../../data/ranges.csv")

bases, ranges
```

```python
import folium
m = folium.Map(location =[30,115], zoom_start = 4)
m
```

```python
for _, base in bases.iterrows():
    # marker
    folium.Marker(
        location = [base["lat"], base["lon"]],
        popup = base["name"],
        tooltip = base["name"]
    ).add_to(m)

    # circles
    for _, r in ranges.iterrows():
        folium.Circle(
            location = [base["lat"], base["lon"]],
            radius = r["radius_km"] * 1000,
            popup = f"{base["name"]} - {r["label"]}",
            color = r["color"],
            fill = True,
            fill_opacity = 0.08
        ).add_to(m)

m
```
